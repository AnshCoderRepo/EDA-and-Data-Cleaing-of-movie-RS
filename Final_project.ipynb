```python
# Initialize empty lists for cleaning
rev = []
for i in df['gross_income'].values:
    rev.append(i)

# Clean the 'gross_income' column
rev_c = []
for i in rev:
    # Convert to string
    i = str(i)
    # Remove commas and dollar signs
    i = i.replace(',', '').replace('$', '')
    # Remove 'M' for million
    if 'M' in i:
        i = i.replace('M', '')
        # Convert to float and multiply by 1 million
        i = float(i) * 1e6
    rev_c.append(i)

# Assign the clean list to the 'gross_income' column
df['gross_income'] = rev_c

# Convert 'gross_income' column to float data type
df['gross_income'] = df['gross_income'].astype(float)

# Check data types of the DataFrame
print(df.dtypes)
```

# # Fixing Votes column¶


```python
# Initialize empty lists for cleaning
votesn = []
for i in df['votes'].values:
    votesn.append(i)

# Clean the 'votes' column
v1 = []
for i in votesn:
    # Convert to string
    i = str(i)
    # Remove commas
    if ',' in i:
        i = i.replace(',', '')
    v1.append(i)

## Convert 'votes' column to numeric data type (float)
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')

# Round the numerical values in the 'votes' column
df['votes'] = df['votes'].round()

# Fill NaN values in the 'votes' column with 0
df['votes'].fillna(0, inplace=True)

# Convert 'votes' column to integer data type
df['votes'] = df['votes'].astype(int)

# Check data types of the DataFrame
print(df.dtypes)



```

# **How many different Genre of Movies are there in this dataset?**

```python
# for unique pair Count
df['genre'].nunique()
```

```python
# Different Unique Genre count
temp=[]
for i in df['genre']:
    d=i.replace(',','').split()# split the combination of genre for single movie
    for j in d:
        temp.append(j)   # add it to temp list and get the unique genre using set
unique_genre = list(set(temp))
len(unique_genre)
```

# **. Which Genre is having most number of Movies in the dataset**

```python
# Movies Count w.r.t unique Genre's Pairs**

a=df.groupby('genre',as_index=False).agg({'id':'count'})
a.sort_values(by='id',ascending=False)
```

```python
df.groupby('genre',as_index=False).count().sort_values(by=['name'],ascending=False)[['genre','name']]
```

```python
# Movies Count w.r.t unique Genre
movies_count = []

for gen in unique_genre:
    cnt=0
    for i in df['genre']:
        if gen in i:
            cnt = cnt+1
    movies_count.append([gen,cnt])

movies_count

df_genre = pd.DataFrame(data=movies_count,columns=['genre_name','cnt'])
df_genre.sort_values(by=['cnt'],ascending=False)
```
